# Dense HEA LeJEPA Features on DRIVE

This notebook pretrains the **dense, teacher-free HEA LeJEPA adaptor** on the same DRIVE retina data used by the segmentation notebook.

What it does:
- loads unlabeled DRIVE images and field-of-view masks
- samples retina patches for dense self-supervised training
- trains `DenseLeJEPAModel` with aligned multi-view augmentations
- extracts dense latent feature maps and pooled feature vectors
- visualizes latent norm maps and a PCA color view of the learned dense features

This is **feature pretraining**, not pseudo-label generation. The model learns dense latent fields with a symmetric LeJEPA objective: dense invariance across aligned views plus SIGReg.

```python
from pathlib import Path
import re
import sys
import random

import matplotlib.pyplot as plt
import numpy as np
import torch
from PIL import Image
from torch.utils.data import DataLoader, Dataset

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == "examples" else Path.cwd().resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from local_conv_attention import DenseLeJEPAModel, load_experiment_config

device = torch.device("mps" if torch.backends.mps.is_available() else ("cuda" if torch.cuda.is_available() else "cpu"))
print("Project root:", PROJECT_ROOT)
print("Device:", device)
torch.manual_seed(0)
np.random.seed(0)
random.seed(0)
```

```python
candidate_roots = [
    PROJECT_ROOT / "data" / "DRIVE",
    PROJECT_ROOT / "examples" / "DRIVE",
]
DATA_ROOT = next((root for root in candidate_roots if root.exists()), None)
assert DATA_ROOT is not None, "Could not find DRIVE under data/DRIVE or examples/DRIVE."
print("Using DRIVE root:", DATA_ROOT)
```

```python
def drive_id_from_name(path: Path) -> str:
    match = re.match(r"(\d+)_", path.stem)
    if match is None:
        raise ValueError(f"Could not parse DRIVE id from {path.name}")
    return match.group(1)


def read_rgb(path: Path) -> np.ndarray:
    image = np.array(Image.open(path).convert("RGB"), dtype=np.float32) / 255.0
    return image


def read_binary_mask(path: Path) -> np.ndarray:
    mask = np.array(Image.open(path), dtype=np.float32)
    if mask.ndim == 3:
        mask = mask[..., 0]
    return (mask > 0).astype(np.float32)


def load_drive_split(split: str) -> list[dict]:
    split_root = DATA_ROOT / split
    image_dir = split_root / "images"
    fov_dir = split_root / "mask"
    samples = []
    for image_path in sorted(image_dir.glob("*.tif")):
        sample_id = drive_id_from_name(image_path)
        fov_path = next(fov_dir.glob(f"{sample_id}_*mask.*"))
        samples.append({"id": sample_id, "image": image_path, "fov_mask": fov_path})
    return samples


train_samples = load_drive_split("training")
test_samples = load_drive_split("test")
all_samples = train_samples + test_samples
print("Training images:", len(train_samples))
print("Test images:", len(test_samples))
print("Unlabeled SSL pool:", len(all_samples))
```

```python
class DriveDenseSSLPatchDataset(Dataset):
    def __init__(
        self,
        samples: list[dict],
        *,
        patch_size: int = 128,
        stride: int = 96,
        min_valid_fraction: float = 0.5,
    ) -> None:
        self.patch_size = patch_size
        self.items: list[dict] = []
        for sample in samples:
            fov = read_binary_mask(sample["fov_mask"])
            height, width = fov.shape
            for y in range(0, height - patch_size + 1, stride):
                for x in range(0, width - patch_size + 1, stride):
                    patch = fov[y:y + patch_size, x:x + patch_size]
                    if patch.mean() >= min_valid_fraction:
                        self.items.append({"sample": sample, "y": y, "x": x})

    def __len__(self) -> int:
        return len(self.items)

    def __getitem__(self, index: int) -> dict[str, torch.Tensor]:
        item = self.items[index]
        image = read_rgb(item["sample"]["image"])
        fov = read_binary_mask(item["sample"]["fov_mask"])
        y = item["y"]
        x = item["x"]
        ps = self.patch_size
        image = image[y:y + ps, x:x + ps]
        fov = fov[y:y + ps, x:x + ps]
        return {
            "image": torch.from_numpy(image.transpose(2, 0, 1)).float(),
            "valid_mask": torch.from_numpy(fov[None]).float(),
            "sample_id": item["sample"]["id"],
        }


PATCH_SIZE = 128
PATCH_STRIDE = 96
ssl_dataset = DriveDenseSSLPatchDataset(
    all_samples,
    patch_size=PATCH_SIZE,
    stride=PATCH_STRIDE,
    min_valid_fraction=0.5,
)
ssl_loader = DataLoader(ssl_dataset, batch_size=4, shuffle=True, drop_last=True)
print("SSL patches:", len(ssl_dataset))

sample = ssl_dataset[0]
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].imshow(sample["image"].permute(1, 2, 0))
axes[0].set_title(f"Patch {sample['sample_id']}")
axes[1].imshow(sample["valid_mask"][0], cmap="gray")
axes[1].set_title("FOV valid mask")
for ax in axes:
    ax.axis("off")
plt.tight_layout()
```

```python
config = load_experiment_config(PROJECT_ROOT / "configs" / "hea_dense_lejepa_default.yaml")
config.model.in_channels = 3
config.model.base_channels = 8
config.model.encoder_depths = [1, 1, 1, 1]
config.model.decoder_depths = [1, 1, 1]
config.model.attention.heads = 2
config.model.attention.head_dim = 8
config.model.latent.latent_dim = 64
config.model.lejepa.num_views = 4
config.model.lejepa.lambda_sigreg = 0.02
config.model.lejepa.invariance.loss_on_valid_only = True
config.model.lejepa.views.mode = "aligned_same_geometry"
config.model.lejepa.views.corruption.random_block_mask = True
config.model.lejepa.views.corruption.block_mask_ratio = 0.2
config.model.lejepa.sigreg.num_slices = 64
config.model.lejepa.sigreg.num_knots = 17
config.validate()

model = DenseLeJEPAModel(config.model).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
num_params = sum(parameter.numel() for parameter in model.parameters())
print(f"Dense LeJEPA parameters: {num_params:,}")
```

```python
def train_dense_lejepa(model, loader, optimizer, *, epochs: int = 5) -> dict[str, list[float]]:
    history = {"loss": [], "inv_loss": [], "sigreg_loss": []}
    for epoch in range(epochs):
        model.train()
        running = {key: 0.0 for key in history}
        batches = 0
        for batch in loader:
            images = batch["image"].to(device)
            valid_mask = batch["valid_mask"].to(device)
            optimizer.zero_grad(set_to_none=True)
            out = model(images, valid_mask=valid_mask)
            out["loss"].backward()
            optimizer.step()
            for key in history:
                running[key] += float(out[key].detach())
            batches += 1
        for key in history:
            history[key].append(running[key] / max(1, batches))
        print(
            f"epoch {epoch + 1:02d} | loss={history['loss'][-1]:.4f} | "
            f"inv={history['inv_loss'][-1]:.4f} | sigreg={history['sigreg_loss'][-1]:.4f}"
        )
    return history


history = train_dense_lejepa(model, ssl_loader, optimizer, epochs=5)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 3))
axes[0].plot(history["loss"], marker="o")
axes[0].set_title("Total loss")
axes[1].plot(history["inv_loss"], marker="o")
axes[1].set_title("Dense invariance")
axes[2].plot(history["sigreg_loss"], marker="o")
axes[2].set_title("SIGReg")
for ax in axes:
    ax.set_xlabel("epoch")
plt.tight_layout()
```

```python
@torch.no_grad()
def extract_dense_latent_map(model, image: torch.Tensor) -> torch.Tensor:
    model.eval()
    image = image.unsqueeze(0).to(device)
    features = model.backbone(image)
    latents = model.projector(features)
    return latents[0].cpu()


def pca_colorize_latents(latent_map: torch.Tensor) -> np.ndarray:
    channels, height, width = latent_map.shape
    flat = latent_map.reshape(channels, -1).T.numpy()
    flat = flat - flat.mean(axis=0, keepdims=True)
    _, _, vh = np.linalg.svd(flat, full_matrices=False)
    coords = flat @ vh[:3].T
    coords = (coords - coords.min(axis=0, keepdims=True)) / (coords.max(axis=0, keepdims=True) - coords.min(axis=0, keepdims=True) + 1e-6)
    return coords.reshape(height, width, 3)


full_image = torch.from_numpy(read_rgb(all_samples[0]["image"]).transpose(2, 0, 1)).float()
full_fov = torch.from_numpy(read_binary_mask(all_samples[0]["fov_mask"])[None]).float()
latent_map = extract_dense_latent_map(model, full_image)
latent_norm = torch.linalg.vector_norm(latent_map, dim=0)
latent_pca = pca_colorize_latents(latent_map)

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
axes[0].imshow(full_image.permute(1, 2, 0))
axes[0].set_title("DRIVE image")
axes[1].imshow(full_fov[0], cmap="gray")
axes[1].set_title("FOV mask")
axes[2].imshow(latent_norm, cmap="magma")
axes[2].set_title("Latent norm")
axes[3].imshow(latent_pca)
axes[3].set_title("Latent PCA color")
for ax in axes:
    ax.axis("off")
plt.tight_layout()
print("Dense latent map shape:", tuple(latent_map.shape))
```

```python
@torch.no_grad()
def build_feature_bank(model, dataset, *, max_items: int = 16) -> tuple[np.ndarray, list[str]]:
    pooled_features = []
    ids = []
    for index in range(min(max_items, len(dataset))):
        sample = dataset[index]
        latent = extract_dense_latent_map(model, sample["image"])
        valid = sample["valid_mask"]
        pooled = (latent * valid).sum(dim=(1, 2)) / valid.sum().clamp_min(1.0)
        pooled_features.append(pooled.numpy())
        ids.append(sample["sample_id"])
    return np.stack(pooled_features), ids


feature_bank, feature_ids = build_feature_bank(model, ssl_dataset, max_items=16)
print("Feature bank shape:", feature_bank.shape)
print("First few sample ids:", feature_ids[:5])

similarity = feature_bank @ feature_bank.T
plt.figure(figsize=(5, 4))
plt.imshow(similarity, cmap="viridis")
plt.title("Patch feature similarity")
plt.colorbar()
plt.tight_layout()
```
