# HEA-UNet Retina Example

This notebook demonstrates **Hierarchical Elevator Attention U-Net (HEA-UNet)** on a standard retina sample image using the modules in this repository.

What this notebook does:

- loads the standard `skimage.data.retina()` sample image
- builds a lightweight pseudo vessel mask with classical image processing
- configures a small HEA-UNet from YAML + Python overrides
- trains on retina patches for a few epochs
- runs tiled inference and visualizes the prediction map

This is a **worked example**, not a scientific benchmark. For real experiments, replace the pseudo mask generation with true labels from datasets such as DRIVE, STARE, or CHASE_DB1.


```python
import os
import random
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import torch
from skimage import data, exposure, filters, morphology
from torch import nn
from torch.utils.data import DataLoader, Dataset

os.environ.setdefault("MPLCONFIGDIR", str(Path("/tmp/mplconfig")))

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == "examples" else Path.cwd().resolve()
sys.path.insert(0, str(PROJECT_ROOT))

from local_conv_attention import HEAUNet, load_experiment_config

plt.rcParams["figure.figsize"] = (14, 4)
plt.rcParams["image.cmap"] = "gray"

seed = 0
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)

device = torch.device(
    "mps"
    if getattr(torch.backends, "mps", None) is not None and torch.backends.mps.is_available()
    else "cuda"
    if torch.cuda.is_available()
    else "cpu"
)
print(f"Using device: {device}")
print(f"Project root: {PROJECT_ROOT}")
```

```python
# Load the standard retina sample and generate a lightweight pseudo vessel mask.
image = data.retina().astype(np.float32) / 255.0
green = exposure.equalize_adapthist(image[..., 1], clip_limit=0.02)
vesselness = filters.frangi(green, sigmas=range(1, 4), black_ridges=False)
mask = vesselness > np.quantile(vesselness, 0.92)
mask = morphology.remove_small_objects(mask, 32)
mask = morphology.binary_closing(mask, morphology.disk(1)).astype(np.float32)

print("Image shape:", image.shape)
print("Pseudo-mask positive fraction:", float(mask.mean()))
```

```python
fig, axes = plt.subplots(1, 4, figsize=(18, 4))
axes[0].imshow(image)
axes[0].set_title("Retina sample")
axes[1].imshow(green)
axes[1].set_title("Enhanced green channel")
axes[2].imshow(vesselness)
axes[2].set_title("Frangi vesselness")
axes[3].imshow(mask)
axes[3].set_title("Pseudo vessel mask")
for ax in axes:
    ax.axis("off")
plt.tight_layout()
```

```python
PATCH_SIZE = 128
PATCH_STRIDE = 96
MAX_TRAIN_PATCHES = 48
MAX_VAL_PATCHES = 16


def collect_patch_coords(mask_array: np.ndarray, patch_size: int, stride: int) -> list[tuple[int, int]]:
    coords: list[tuple[int, int]] = []
    height, width = mask_array.shape
    for y in range(0, height - patch_size + 1, stride):
        for x in range(0, width - patch_size + 1, stride):
            patch_mask = mask_array[y : y + patch_size, x : x + patch_size]
            if patch_mask.mean() > 0.01:
                coords.append((y, x))
    return coords


all_coords = collect_patch_coords(mask, PATCH_SIZE, PATCH_STRIDE)
split_x = int(image.shape[1] * 0.65)
train_coords = [coord for coord in all_coords if coord[1] < split_x][:MAX_TRAIN_PATCHES]
val_coords = [coord for coord in all_coords if coord[1] >= split_x][:MAX_VAL_PATCHES]

print(f"Train patches: {len(train_coords)}")
print(f"Val patches:   {len(val_coords)}")


class RetinaPatchDataset(Dataset):
    def __init__(self, image_array: np.ndarray, mask_array: np.ndarray, coords: list[tuple[int, int]], patch_size: int) -> None:
        self.image_array = image_array
        self.mask_array = mask_array
        self.coords = coords
        self.patch_size = patch_size

    def __len__(self) -> int:
        return len(self.coords)

    def __getitem__(self, index: int) -> tuple[torch.Tensor, torch.Tensor]:
        y, x = self.coords[index]
        image_patch = self.image_array[y : y + self.patch_size, x : x + self.patch_size]
        mask_patch = self.mask_array[y : y + self.patch_size, x : x + self.patch_size]
        image_tensor = torch.from_numpy(image_patch.transpose(2, 0, 1)).float()
        mask_tensor = torch.from_numpy(mask_patch[None]).float()
        return image_tensor, mask_tensor


train_loader = DataLoader(
    RetinaPatchDataset(image, mask, train_coords, PATCH_SIZE),
    batch_size=4,
    shuffle=True,
)
val_loader = DataLoader(
    RetinaPatchDataset(image, mask, val_coords, PATCH_SIZE),
    batch_size=4,
    shuffle=False,
)

sample_images, sample_masks = next(iter(train_loader))
print("Train batch image shape:", tuple(sample_images.shape))
print("Train batch mask shape:", tuple(sample_masks.shape))
```

```python
# Start from the YAML preset and then override a few fields for this small notebook demo.
config = load_experiment_config(PROJECT_ROOT / "configs" / "hea_unet_small.yaml")
config.model.in_channels = 3
config.model.num_classes = 1
config.model.base_channels = 8
config.model.encoder_depths = [1, 1, 1, 1]
config.model.decoder_depths = [1, 1, 1]
config.model.attention.heads = 2
config.model.attention.head_dim = 8
config.model.hea.enabled_decoder_stages = [0, 1]
config.model.attention.operator_backend = "optimized"
config.validate()

model = HEAUNet(config.model).to(device)
criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

num_params = sum(parameter.numel() for parameter in model.parameters())
print(f"Model parameters: {num_params:,}")
print(model.config)
```

```python
def dice_score_from_logits(logits: torch.Tensor, targets: torch.Tensor, threshold: float = 0.5) -> float:
    probs = torch.sigmoid(logits)
    preds = (probs > threshold).float()
    intersection = (preds * targets).sum(dim=(1, 2, 3))
    union = preds.sum(dim=(1, 2, 3)) + targets.sum(dim=(1, 2, 3))
    dice = (2.0 * intersection + 1e-6) / (union + 1e-6)
    return float(dice.mean().item())


def run_epoch(model: nn.Module, loader: DataLoader, optimizer: torch.optim.Optimizer | None = None) -> tuple[float, float]:
    is_train = optimizer is not None
    model.train(is_train)
    total_loss = 0.0
    total_dice = 0.0

    for images, targets in loader:
        images = images.to(device)
        targets = targets.to(device)

        if is_train:
            optimizer.zero_grad(set_to_none=True)

        with torch.set_grad_enabled(is_train):
            logits = model(images)
            loss = criterion(logits, targets)
            if is_train:
                loss.backward()
                optimizer.step()

        total_loss += float(loss.item()) * images.size(0)
        total_dice += dice_score_from_logits(logits.detach(), targets.detach()) * images.size(0)

    count = len(loader.dataset)
    return total_loss / count, total_dice / count


history = {"train_loss": [], "train_dice": [], "val_loss": [], "val_dice": []}
EPOCHS = 3

for epoch in range(EPOCHS):
    train_loss, train_dice = run_epoch(model, train_loader, optimizer)
    val_loss, val_dice = run_epoch(model, val_loader, optimizer=None)
    history["train_loss"].append(train_loss)
    history["train_dice"].append(train_dice)
    history["val_loss"].append(val_loss)
    history["val_dice"].append(val_dice)
    print(
        f"Epoch {epoch + 1}/{EPOCHS} | "
        f"train loss={train_loss:.4f}, train dice={train_dice:.4f} | "
        f"val loss={val_loss:.4f}, val dice={val_dice:.4f}"
    )
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(history["train_loss"], marker="o", label="train")
axes[0].plot(history["val_loss"], marker="o", label="val")
axes[0].set_title("BCE loss")
axes[0].set_xlabel("Epoch")
axes[0].legend()

axes[1].plot(history["train_dice"], marker="o", label="train")
axes[1].plot(history["val_dice"], marker="o", label="val")
axes[1].set_title("Dice score")
axes[1].set_xlabel("Epoch")
axes[1].legend()
plt.tight_layout()
```

```python
def predict_full_image(model: nn.Module, image_array: np.ndarray, patch_size: int = 128, stride: int = 96) -> np.ndarray:
    model.eval()
    height, width, _ = image_array.shape
    accum = np.zeros((height, width), dtype=np.float32)
    counts = np.zeros((height, width), dtype=np.float32)

    ys = list(range(0, max(height - patch_size + 1, 1), stride))
    xs = list(range(0, max(width - patch_size + 1, 1), stride))
    if ys[-1] != height - patch_size:
        ys.append(height - patch_size)
    if xs[-1] != width - patch_size:
        xs.append(width - patch_size)

    with torch.no_grad():
        for y in ys:
            for x in xs:
                patch = image_array[y : y + patch_size, x : x + patch_size]
                patch_tensor = torch.from_numpy(patch.transpose(2, 0, 1)).float()[None].to(device)
                logits = model(patch_tensor)
                probs = torch.sigmoid(logits)[0, 0].cpu().numpy()
                accum[y : y + patch_size, x : x + patch_size] += probs
                counts[y : y + patch_size, x : x + patch_size] += 1.0

    return accum / np.maximum(counts, 1.0)


prob_map = predict_full_image(model, image, patch_size=PATCH_SIZE, stride=PATCH_STRIDE)
pred_mask = (prob_map > 0.5).astype(np.float32)
print("Prediction map shape:", prob_map.shape)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(18, 4))
axes[0].imshow(image)
axes[0].set_title("Input retina image")

axes[1].imshow(mask)
axes[1].set_title("Pseudo target")

axes[2].imshow(prob_map, vmin=0.0, vmax=1.0)
axes[2].set_title("HEA-UNet probability")

overlay = image.copy()
overlay[..., 0] = np.maximum(overlay[..., 0], pred_mask)
axes[3].imshow(overlay)
axes[3].set_title("Prediction overlay")

for ax in axes:
    ax.axis("off")
plt.tight_layout()
```
